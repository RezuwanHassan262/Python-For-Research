1. Print your name 50 times

3. Take n and m as input and print all the odd numbers from n to m using Loop.

```python
#Using while loop
n = int(input("Insert a number: "))
m = int(input("Insert another number that is less than before: "))
p = m
while m<n:
  if m%2!=0:
    print(m, end = ' ')
  m+= 1

print('\n')
print("For loop output starts from here")

m = p
#Using for loop
for p in range(m,n,1):
  if p%2!=0:
    if p >= m:
      print(p, end = ' ')
```
