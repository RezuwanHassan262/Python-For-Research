1. Print n to 1 for m times. Here, you need to take n and m as input.

````python
n,m = map(int, input("Insert the number that should be printed upto 1 and the number of times it should be printed respectively: ").split())
if n <0:
  for k in range(1,m+1):
    for i in range(n,2,1):
      print(i,end =' ')
    print('\n')
elif n >0:
  for k in range(1,m+1):
    for i in range(n,0,-1):
      print(i,end =' ')
    print('\n')
else:
  for k in range(1,m+1):
    for i in range(n,2,1):
      print(i, end = ' ')
    print('\n')
````

2. print 1 to n and if find 5 then break

````python
n = int(input("Insert a number: "))
i = 1
while n >= i:
  if i == 5:
    break
  else:
    print(i, end = ' ') 
  i+=1
print('\n')
print("The loop broke at 5. Hence, No more printing")
````
