```python
a = int(input("Insert a number: "))

if (a%7==0 and a%5 ==0):
  print("the number is both divisable by 7 and multiple of 5")
elif(a%7==0 and a%5 !=0):
  print("the number is divisable by 7 but not multiple of 5")
elif(a%7!=0 and a%5 ==0):
  print("the number is not divisable by 7 but multiple of 5")
else:
  print("the number is not divisable by 7, nor multiple of 5")
```
