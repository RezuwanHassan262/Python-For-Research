#**Dummy Encoding**


```python
import category_encoders as ce
import pandas as pd
data=pd.DataFrame({'Elements':['Fire','Water','Earth','Air','Fire','Water','Earth','Air','Air','Air']})

print(data, "\n") #Original Data

data_encoded=pd.get_dummies(data=data,drop_first=True) #drop_first=True because Dummy encoding uses N-1 features to represent N labels/categories.
data_encoded
```

#**Effect Encoding**

Dummy encoding

```python
import category_encoders as ce
import pandas as pd
data=pd.DataFrame({'City':['Delhi','Mumbai','Hydrabad','Chennai','Bangalore','Delhi','Hydrabad','Bangalore','Delhi']})

#Create object for one-hot encoding
encoder=ce.OneHotEncoder(cols='City',handle_unknown='return_nan',return_df=True,use_cat_names=True)

print(data, "\n") #Original Data

data_encoded = encoder.fit_transform(data) #Fit and transform Data
data_encoded
```

Effect encoding

```python
import category_encoders as ce
import pandas as pd
data=pd.DataFrame({'City':['Delhi','Mumbai','Hyderabad','Chennai','Bangalore','Delhi','Hyderabad']})
encoder=ce.sum_coding.SumEncoder(cols='City',verbose=False,)


print(data, "\n") #Original Data

encoder.fit_transform(data)
```

#**Hash Encoding**

```python
import category_encoders as ce
import pandas as pd

#Create the dataframe
data=pd.DataFrame({'Month':['January','April','March','April','Februay','June','July','June','September']})

print(data, "\n") #Original Data

#Create object for hash encoder
encoder=ce.HashingEncoder(cols='Month',n_components=7) #Hashing is the transformation of arbitrary size input in the form of a fixed-size value

encoder.fit_transform(data) #Fit and Transform Data
```

#**Binary Encoding**

```python
import category_encoders as ce
import pandas as pd

#Create the Dataframe
data=pd.DataFrame({'Digit':['Delhi','Mumbai','Hyderabad','Chennai','Bangalore','Delhi','Hyderabad','Mumbai','Agra']})

#Create object for binary encoding
encoder= ce.BinaryEncoder(cols=['Digit'],return_df=True)

print(data, "\n") #Original Data

#Fit and Transform Data 
data_encoded = encoder.fit_transform(data) 
data_encoded
```

#**BaseN Encoding**

```python
import category_encoders as ce
import pandas as pd

#Create the dataframe
data=pd.DataFrame({'City':['Delhi','Mumbai','Hyderabad','Chennai','Bangalore','Delhi','Hyderabad','Mumbai','Agra']})

#Create an object for Base N Encoding
encoder= ce.BaseNEncoder(cols=['City'],return_df=True,base=3) #Changes base according to the base argument

print(data, "\n") #Original Data

#Fit and Transform Data
data_encoded=encoder.fit_transform(data)
data_encoded
```

#**Target Encoding**

```python
import pandas as pd
import category_encoders as ce

#Create the Dataframe
data=pd.DataFrame({'class':['A','B','C','B','C','A','A','A'],'Marks':[50,30,70,80,45,97,80,68]})

#Create target encoding object
encoder=ce.TargetEncoder(cols='class') 

print(data, "\n") #Original Data

#Fit and Transform Train Data
encoder.fit_transform(data['class'],data['Marks'])
```
