7. Write a Python program to remove an item from a set if it is present in the set.

```python
#Removing item not knowing if it is present in the set or not
s = set(range(0,10))
print('Initial set:',s)
s.remove(5)
print('New set:',s)

print('\n')
#Removing item from a set by checking
print('\n')
s = set(range(0,10))
print('Initial set:',s)
l = list(s)
for i in l:
  if i == 5:
    l.pop(i)
  else:
    pass
s = set(l)
print('New set:',s)
```

9. Write a Python script to merge two Python dictionaries.

```python
#Merging two dictionaries
print('Merging two dictionaries')
print('\n')
d1 = {'a': 12, 'b': 5, 'c': 75}
d2 = {'d' : 1, 'e' : 42, 'f' : 100}
d3 = {**d1,**d2}
print("Dictionary 1:", d1)
print("Dictionary 2:", d2)
print("Merged dictionary:", d3)
```

10. Write a Python code that convert a dictionary to two list of keys and values.

```python
l1 = []
l2 = []
d = {'money': 12, 'candy': 5, 'tissues': 75}
print('Actual Dictionary:',d)
for (k,v) in d.items():
  l1.append(k)
  l2.append(v)
print('Keys:',l1)
print('Values:',l2)
```
