#**Linear regression for single features**
#Rainfall with respect to year

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split,cross_val_score

x = data["YEAR"].values.reshape(-1,1)
y = data["AR"].values.reshape(-1,1)

x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.4) #Spliting the dataset. 40% for test purpose, 60% for training purpose and assigning them to a variable
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics

linear_model = LinearRegression() #Create new linear model object
linear_model.fit(x_train,y_train) #Fitting the training dataset to train the model
```

```python
y_predict = linear_model.predict(x_test) #After training the model, Giving the "x_test" as input to predict the output 
print(y_predict) #Printing the predicted output
```

```python
print("Intercept, C =",linear_model.intercept_) #To retrieve the intercept
print("Slope/Co-efficient, M =",linear_model.coef_) #To retrieve the slope
```

```python
x = [[2020],[2021],[2022]] #Prediction of rain for 2020,2021,2022
linear_model.predict(x)
```

```python
df = pd.DataFrame({"Actual":y_test.flatten(),"Predicted":y_predict.flatten()})
df
```

```python
df1 = df.head(25)
df.plot(kind = 'bar',figsize = (10,10))
plt.grid(which = 'major',linestyle = '-',linewidth = '0.5',color = 'green')
plt.grid(which = 'minor',linestyle = ':',linewidth = '0.5',color = 'black')
plt.xlabel("Year")
plt.ylabel("Rainfall")
plt.title("Year VS Rainfall for Dinajpur")
plt.show()
```

```python
plt.plot(df)
```

```python
print("Mean absolute error:",metrics.mean_absolute_error(y_test,y_predict))
print("Mean squared error:",metrics.mean_squared_error(y_test,y_predict))
print("Mean absolute error:",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```


#Rice prediction for different features

```python
x = data[['YEAR','AR','AMT','AMiT','AH','R Area']].values
y = data['R Pro'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
df = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df.head(5)
```

```python
df.plot(kind='bar',figsize = (10,6))
plt.grid(which='major',linestyle='-',linewidth='0.5',color='green')
plt.grid(which='minor',linestyle=':',linewidth='0.5',color='black')
plt.plot(df,linewidth=4)
plt.show()
```

```python
print("Mean absolute error:",metrics.mean_absolute_error(y_test,y_pred)/10000)
print("Mean squared error:",metrics.mean_squared_error(y_test,y_pred)/(10000*10000))
print("Mean absolute error:",np.sqrt(metrics.mean_squared_error(y_test,y_pred))/10000)
```
