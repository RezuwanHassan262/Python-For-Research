##**1. Design All the plot that are Design by Matplotlib using Seaborn.**

Importing seaborn & other relevant libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

1a. Loading dataset for line plot

```python
x  = sns.load_dataset("anagrams")
x
```

1b. Lineplot using seaborn on the loaded dataset


```python
df = x[["num1", "num2"]]
sns.lineplot(data = df)
plt.xticks((0.0,2.5,5.0,7.5,10.0,12.5,15.0,17.5),(range(0,22,3)))
plt.legend(labels = ["num1","num2"],loc =2)
plt.xlabel("subidr")
plt.ylabel("values")
plt.grid()
```

2a. Loading dataset for scatter plot

```python
x  = sns.load_dataset("iris")
x
```

2b. Scatterplot using seaborn on the loaded dataset

```python
df = x[["sepal_length", "sepal_width"]]
sns.lmplot('sepal_length', 'sepal_width', data=df, fit_reg=False)
plt.grid()
```

3a. Loading dataset for histogram

```python
z  = sns.load_dataset("tips")
z
```

3b. histogram using seaborn on the loaded dataset


```python
df = z[["sex"]]
plt.hist(df.sex, alpha=.3)
sns.rugplot(df.sex)
plt.ylabel("Number of people")
plt.grid()
```

4a. Loading dataset for bar plot

```python
z  = sns.load_dataset("titanic")
z.head()
```

4b. barplot using seaborn on the loaded dataset


```python
sns.barplot(x = 'class', y = 'fare', data = z)
plt.grid()
plt.show()
```

5a. Loading dataset for Pie chart

```python
z  = sns.load_dataset("diamonds")
z.head()
```

5b. Pie chart using seaborn on the loaded dataset


```python
#Grouping the cut by price
data = z.groupby("cut")["price"].sum()
l = list(data.keys())
data
```

```python
colors = sns.color_palette(['#009ACD', '#ADD8E6', '#63D1F4', '#0EBFE9','#C1F0F6'])[0:5]
plt.pie(data, labels = l, colors = colors, autopct='%.2f%%')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

x = pd.read_excel('Assignment 12 DataSet.xlsx')
df = pd.DataFrame(x)
df
```

**1. Histogram**

```python
p = df[["state"]]
plt.hist(p.state, alpha=.3)
plt.title("Frequency of states in this dataset")
plt.xlabel("States")
plt.ylabel("Frequency")
plt.grid()
```

**2. Box plot**

```python
s = df["state"].to_list()
p1 = df["postcode"].to_list()
p2 = df["property_valuation"].to_list()
plt.bar(s,p1,label = "postcode", width = 0.5)
plt.bar(s,p2,label = "property_valuation", width = 0.5)
plt.xlabel("States")
plt.ylabel ("Quantity")
plt.legend(loc = "best")
plt.grid()
#plt.axis([-2,5,0,100])
```

**3. Pie chart**

```python
s = df["state"].to_list()
d = {}
for i in s:
  if i not in d:
    d[i] = 1
  else:
    d[i] = d[i]+1
print("Keys & Frequency: ",d,"\n")
k = d.keys()
v = d.values()
plt.pie(v,labels = k,autopct= "%.2f%%")
plt.title("Frequency of states present in this dataset")
```

**4. Scatter plot**

```python
p1 = df["postcode"].to_list()
p2 = df["property_valuation"].to_list()
plt.scatter(p1,p2)
plt.xlabel("Postcodes")
plt.ylabel("Property Value")
plt.title("postcode Vs. property value")
plt.grid()
```
