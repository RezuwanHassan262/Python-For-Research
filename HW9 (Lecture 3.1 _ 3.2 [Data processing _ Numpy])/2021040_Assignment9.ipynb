4. Find the elements that are multiple of 3 as well as divisable by 5 from a numpy array.

```python
import numpy as np

#Technique 1

a = np.arange(0,50)
print('array =', a)
print('\n')
i = 0
b = len(a)
print('Technique 1 output')
print('\n')
while i <= b:
  if (3*i in a) and ((3*i)%5==0):
    print(3*i, end =' ')
  i+=1
print('\n')

#Technique 2

a = np.arange(0,50)
print('array =', a)
print('\n')
print('Technique 2 output')
print('\n')
for i in a:
  if (i%3 == 0) and (i%5==0):
    print(i, end =' ')
```

5. Create a new array from a given array, swapping first and last, second and third columns

```python
import numpy as np

#Swapping columns
print("Swappping columns",'\n')
a = np.random.randint(0,20,(4,4))
print('initial array','\n')
print(a)
print('new sorted array','\n')
a[:,[0,3]]=a[:,[3,0]]
a[:,[1,2]]=a[:,[2,1]]
print(a)

#Swapping rows
print("Swappping rows",'\n')
a = np.random.randint(0,20,(4,4))
print('initial array','\n')
print(a)
print('new sorted array','\n')
b = a[0]
c = a[1]
d = a[2]
e = a[3]
a = np.concatenate((e,d,c,b))
a = a.reshape(4,4)
print('new sorted array','\n')
print(a)
```

8. Take a 2D numpy array and sort it in assending and dissending order.

```python
import numpy as np

a = np.random.randint(0,20,(4,5))
print('initial matrix','\n')
print(a,'\n')

b = a[0]
c = a[1]
d = a[2]
e = a[3]
f = np.concatenate((b,c,d,e))
g = np.sort(f)
i = g.reshape(4,5)
print('New array in ascending order:','\n')
print(i)
print('\n')
h = g[::-1]
j = h.reshape(4,5)
print('New array in dscending order:','\n')
print(j)
```

9. Convert a Python dictionary to Numpy array.

```python
import numpy as np

#Technique 1

d = {'a':6, 'b':5, 'c':4, 'd':3, 'e':2, 'f':1}
l = list(d.items())
x = []
for i in l:
  l =list(i)
  x = np.concatenate((x,l))
print('Technique 1 output','\n')
x = x.reshape(6,2)
print(x)

#Technique 2

d = {'a':6, 'b':5, 'c':4, 'd':3, 'e':2, 'f':1}
l = list(d.items())
x = np.array(l)
print('\n')
print('Technique 2 output','\n')
print(x)
```

10. Write a code that show all the numpy array generation procedure.

```python
import numpy as np
import matplotlib.pyplot as plt

print('np.arange','\n')
a = np.arange(0,10)
print(a,'\n')

print('multidimensional np.arange','\n')
a = np.arange(0,10).reshape(2,5)
print(a,'\n')

print('\n')

print('randomly array creation','\n')
a = np.random.randint(0,10,10)
print(a,'\n')

print('multidimensional random array creation','\n')
a = np.random.randint(0,10,10).reshape(2,5)
print(a,'\n')

print('\n')

l = list(range(0,10))
print('list =',l,'\n')
print('converting a list into array','\n')
a = np.array(l)
print(a,'\n')

print('converting a list into a multidimentional array','\n')
a = np.array(l).reshape(2,5)
print(a,'\n')

print('\n')

print('only zeros','\n')
a = np.zeros(6)
print(a,'\n')

print('\n')

print('only ones','\n')
a = np.ones(6)
print(a,'\n')

print('\n')

print('generating an identity matrix','\n')
a = np.eye(3)
print(a,'\n')

print('\n')

print('Generating customized diagonal identity matrix','\n')
a = np.diag([1,2,5])
print(a,'\n')

print('\n')

print('using linspace','\n')
a = np.linspace(0,10,5)
print(a,'\n')
print('\n')

print('using logspace','\n')
x = np.linspace(1,2,5)
print('linspace_array,x =',x,'\n')
y = np.logspace(1,2,5)
print('logspace_array,x =',y,'\n')
```
